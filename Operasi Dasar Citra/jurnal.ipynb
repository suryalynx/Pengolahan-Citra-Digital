MODUL 2 OPERASI DASAR CITRA

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

fungsi diatas merupakan fungsi untuk memanggil library yang sudah ada dalam python,
cv2 library python yang berfungsi untuk membaca, menampilkan, dan menyimpan gambar,
matplotlib library python yang berfungsi untuk memvisualisasikan data,
numpy library python yang berfungsi untuk melakukan operasi vektor dan matriks dengan mengolah array multidimensi.

```python
image = cv2.imread("beanny.jpg") 
gray = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
plt.imshow(gray,cmap="gray")
```

pada fungsi diatas merupakan fungsi untuk mengubah atau konversi warna gambar yang awalnya RGB menjadi GRAY dengan fungsi yang sudah disediakan oleh library cv2, kemudian menampilkan gambar tersebut dengan fungsi yang sudah disediakan oleh library matplotlib.

```python
tinggi, panjang = gray.shape
```

pada potongan code diatas berfungsi untuk tinggi, panjang = gray.shape untuk mengetahui ukuran dari tinggi dan panjang dari citra tersebut.


NOMOR 1

Lakukan operasi dibawah ini menggunakan gambar diatas:

TRANSLASI

 translasi sejauh x = -50 dan y = 40

```python
translasi = np.zeros_like(gray)
x_shift = -50
y_shift = 40
for i in range(tinggi):
    for j in range(panjang):
        if (i - y_shift) >= 0 and (i - y_shift) < tinggi and (j - x_shift) >= 0 and (j - x_shift) < panjang:
            translasi[i][j] = gray[i - y_shift][j - x_shift]
plt.imshow(translasi, cmap='gray')
```

pada fungsi diatas merupakan fungsi translasi yang digunakan untuk menggeser sebuah citra sesuai dengan yang diinginkan. pada translasi citra ini akan menggeser sejauh x=-50 dan y = 40. langkah awal yang dilakukan adalah membuat matriks yang berukuran sama dengan citra tersebut, menentukan nilai x dan y yang akan digeser, kemudian melakukan perulangan untuk mengambil nilai tersebut sehingga dapat dilakukan translasi sejauh x=-50 dan y=40

ROTASI

Rotasi dengan sudut 70 derajat

```python
rotasi = np.zeros_like(gray)
angle = 70*(np.pi/180)
tengahX = 435 /2
tengahY = 435 /2
for i in range(435):
    for j in range(435):
        x2 = int((i - tengahX) * np.cos(angle) - (j - tengahY) * np.sin(angle) + tengahX)
        y2 = int((i - tengahX) * np.sin(angle) + (j - tengahY) * np.cos(angle) + tengahX)
        if 0 <= x2 < 435 and 0 <= y2 < 435:
           rotasi[i][j] = gray[x2][y2]
plt.imshow(rotasi, cmap="gray")
```

pada fungsi diatas merupakan fungsi untuk melakukan rotasi pada citra sejauh yang diinginkan. pada citra ini akan dirotasikan sejauh 70 derajat searah jarum jam. langkah awal yang dilakukan adalah membuat matriks yang sama dengan ukuran citra tersebut, kemudian menentukan sudut rotasi dalam derajat, lalu mengonversinya menjadi radian. setelah itu akan dilakukan proses menjadi titik tengah dari kedua sumbu x dan y dengan cara bagi dua dari ukuran citra tersebut, kemudian akan melakukan perulangan untuk membuat citra berotasi sesuai yang diinginkan dengan rumus yang sudah ada dengan perulangan tersebut. langkah terakhir adalah menampilkan hasil rotasi citra tersebut.

PENCERMINAN

Penceriman terhadap sumbu x dan sumbu y

```python
cermin = np.zeros(gray.shape)
x=panjang-1
```

membuat sebuah array NumPy baru dengan nama cermin yang memiliki ukuran yang sama dengan citra grayscale asli (gray), dan diinisialisasi dengan nilai nol menggunakan fungsi np.zeros(). kemudian membuat variabel x yang berisi nilai lebar citra panjang-1. Hal ini dilakukan untuk menghindari index out-of-range pada saat melakukan operasi citra.

SUMBU X

```python
for i in range(panjang):
    for j in range(panjang):
        cermin[i][j]=gray[x][j]
    x-=1
```

pada fungsi ini merupakan fungsi untuk melakukan flip atau pemcerminan secara horizontal pada citra dengan menggunakan fungsi perulangan dengan proses didalamnya merupakan proses flip dari citra tersebut dengan metode flip horizontal atau dengan sumbu x pada citra tersebut

```python
plt.imshow(cermin, cmap="gray")
```

pada fungsi diatas merupakan fungsi untuk menampilkan hasil dari percerminan sumbu x atau flip horizontal dari citra tersebut

SUMBU Y

```python
for i in range(panjang):
    for j in range(panjang):
        cermin[j][i]=gray[j][x]
    x-=1
```

pada fungsi ini merupakan fungsi untuk melakukan flip atau pemcerminan secara vertikal pada citra dengan menggunakan fungsi perulangan dengan proses didalamnya merupakan proses flip dari citra tersebut dengan metode flip vertikal atau dengan sumbu y pada citra tersebut

```python
plt.imshow(cermin, cmap="gray")
```

pada fungsi diatas merupakan fungsi untuk menampilkan hasil dari percerminan sumbu y atau flip vertikal dari citra tersebut

DILATASI

Dilatasi matriks dengan faktor skala = 3

```python
skala = 3
dilatasi = np.zeros((tinggi*skala, panjang*skala))
```

pada potongan code diatas merupakan sebuah variabel yang akan digunakan padad proses dilatasi. skala = 3 merupakan variabel untuk membuat faktor skalanya = 3 dan dilatasi merupakan variabel yang digunakan untuk menyimpan matriks kosong dengan tinggi dan panjang dari citra tersebut akan dikalian dengan skala yaitu 3.

```python
for i in range(len(dilatasi)):
    for j in range(len(dilatasi)):
      dilatasi[i][j] = gray[int(i/3),int(j/3)]
plt.imshow(dilatasi, cmap="gray")
```

pada fungsi diatas merupakan fungsi untuk melakukan perulangan untuk menghasilkan dilatasi dari citra tersebut yaitu dengan melakukan perulangan mengecek setiap baris dan kolom pada citra kemudian melakukan proses dilatasi didalamnya dengan  membagi nilai i dan j dengan skala 3 sehingga akan menghasilkan index pada citra asli yang lebih kecil

NOMOR 2

Lakukan operasi citra negatif dari citra graysclae lalu bandingkan histogram kedua citra tersebut.

```python
negatif=np.zeros((435,435)).astype(int)
bar = [x for x in range (256)]
```

Pada baris pertama, kita membuat sebuah array NumPy kosong negatif dengan ukuran (435, 435) dan tipe data integer dengan menggunakan np.zeros() dan astype().

Pada baris kedua, kita membuat sebuah list bar dengan menggunakan list comprehension. List ini berisi 256 angka berturut-turut dari 0 hingga 255. List ini digunakan sebagai representasi nilai piksel dalam citra dengan tipe data unsigned integer 8-bit, yang berkisar dari 0 hingga 255.

```python
for i in range (panjang):
    for j in range (panjang):
        negatif[i][j] =255 - gray[i][j]
```

pada fungsi diatas merupakan fungsi untuk membuat citra menjadi negatif dengan melakukan fungsi perulangan pada citra dengan mengetahui panjang dari citra tersebut kemudian dilakukan pengolahan citra yang awalnya gray di invers menjadi negatif

```python
plt.subplot(1,2,1)
plt.imshow(gray, cmap='gray')

plt.subplot(1,2,2)
plt.imshow(negatif, cmap='gray')
```

Pada fungsi diatas merupakan fungsi untuk menampilkan citra gray dan negatif setelah dilakukan pengolahan atau pemrosesan pada citra tersebut, sehingga dapat dilihat perbedaan dari citra yang awalnya citra masih grayscale kemudian diubah menjadi negatif dengan proses-proses yang telah dilakukan sebelumnya.

```python
normal = np.zeros(256).astype(int)
for i in range (panjang):
    for j in range (panjang):
        normal[gray[i][j]] +=1
```

Pada fungsi diatas digunakan untuk melalukan fungsi perulangan untuk membuat histogram dari citra gray. langkah awal yang dilakukan adalah membuat matriks kosong kemudian mengolahnya ke dalam fungsi perulangan yang telah dibuat sehingga akan menghasilkan hasil histogram dari citra gray tersebut.

```python
plt.bar(bar, normal)
```

Pada fungsi diatas fungsi yang digunakan untuk membuat histogram dari citra yang telah diubah warnanya menjadi gray sehingga setelah dilalukan proses membuat histogram hasil histogram dari citra tersebut akan menjadi seperti histogram di atas.

```python
invers = np.zeros(256).astype(int)
for i in range (panjang):
    for j in range (panjang):
        invers[negatif[i][j]] +=1
```

Pada fungsi diatas digunakan untuk melalukan fungsi perulangan untuk membuat histogram dari citra invers. langkah awal yang dilakukan adalah membuat matriks kosong kemudian mengolahnya ke dalam fungsi perulangan yang telah dibuat sehingga akan menghasilkan hasil histogram dari citra invers tersebut.

```python
plt.bar(bar, invers)
```

Pada fungsi diatas merupakan fungsi untuk menampilkan hasil histogram dari citra gambar yang sudah menjadi negatif atau citra setelah dilakukan proses invers
