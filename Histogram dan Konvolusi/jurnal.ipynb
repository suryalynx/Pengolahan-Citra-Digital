1. Berdasarkan gambar berikut, munculkanlah diagram dan nilai histogram skala warnanya berdasarkan probabilitas piksel, dan jumlah warna keabuan

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

fungsi diatas merupakan fungsi untuk memanggil library yang sudah ada dalam python,
cv2 library python yang berfungsi untuk membaca, menampilkan, dan menyimpan gambar,
matplotlib library python yang berfungsi untuk memvisualisasikan data,
numpy library python yang berfungsi untuk melakukan operasi vektor dan matriks dengan mengolah array multidimensi.

```python
img = cv2.imread('cipung.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

pada fungsi tersebut , cv2 untuk membaca gambar dan mengubah atau konversi warna gambar dari BGR menjadi RGB dan matplotlib dengan nama plt akan menampilkan gambar beserta ukuran gambar tersebut

```python
cipung = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(cipung, cmap='gray')
```

pada fungsi diatas merupakan fungsi untuk mengubah atau konversi warna gambar yang awalnya RGB menjadi GRAY dengan fungsi yang sudah disediakan oleh library cv2, kemudian menampilkan gambar tersebut dengan fungsi yang sudah disediakan oleh library matplotlib.

```python
tinggi, lebar = cipung.shape
print('tinggi :',tinggi,'lebar :',lebar)
```

fungsi diatas merupakan fungsi untuk mencari tinggi dan lebar pada citra, pada library numpy terdapat fungsi shape yang berfungsi menghasilkan jumlah baris dan kolom. pada fungsi print(tinggi , lebar) berfungsi untuk menampilkan hasil tinggi dan lebar dari citra tersebut

```python
hitungAbu = np.zeros(256)
```

menginisialisasikan variabel hitungAbu dengan membuat matriks 0 sebanyak 256 element 

```python
hitungAbu = np.zeros(256).astype(int)
```

fungsi ini merupakan fungsi untuk mencari intensitas keabuan disetiap pixel citra tersebut. dengan bantuan library numpy untuk membuat matriks yang berisi 0 sebanyak 256 element dengan tipe data int

```python
for i in range (tinggi):
    for j in range (lebar):
        # menyimpan nilai dari perulangan yang telah dilakukan ke dalam variabel nilaiAbu
        nilaiAbu = cipung[i,j]
        #menyimpan nilai nilaiAbu pada hitungAbu dengan ditambah 1 
        hitungAbu[nilaiAbu]+= 1
```

pada fungsi perulangan tersebut terdapat beberapa proses yang pertama adalah melakukan perulangan mengecek pixel secara horizontal dengan menghitung tinggi dan melukan perulangan mengecek pixel secara vertikal.

```python
print(hitungAbu)
```

pada fungsi diatas merupakan fungsi untuk menampilkan hasil dari hitungAbu yang telah dibuat sebelumnya.

```python
print(nilaiAbu)
```

pada fungsi diatas merupakan fungsi untuk menampilkan hasil dari nilaiAbu yang telah dibuat sebelumnya.

```python
hitungAbu.sum()
```

pada fungsi diatas merupakan fungsi untuk menampilkan hasil dari tinggi yang dikali dengan lebar dan disimpan pada variabel ni.

```python
# untuk mencari nilai keseluruhan dari nilai keabuan pada citra
histogram = [i for i in range (256)]
# membuat judul pada histogram
plt.title('histogram keabuan')
plt.xlabel('level keabuan')
plt.ylabel('tinggi kecerahan')
# menampilkan histogram dengan membandingkan nilai variabel histogram dengan variabel hitungAbu
plt.bar(histogram, hitungAbu)
```

pada fungsi diatas merupakan fungsi untuk menampilkan histogram keabuan dari gambar cipung yang telah dimasukkan sebelumnya.

```python
plt.title('histogram probabilitas')
plt.bar(histogram, hitungAbu/(tinggi*lebar))
```

pada fungsi diatas merupakan fungsi untuk menghitung probabilitas dari warna keabuan pada gambar cipung dengan menggunakan library matplotlib dengan fungsi bar untuk membuat histogram seperti gambar diatas.

2. Lakukanlah konvolusi pada citra dengan zero padding dan tanpa padding dengan kernel berikut
     ([[0, 0,-1, 1],
        [0, 2, 1, -1],
        [-1, 1, 2, 0],
        [1, -1, 0, 0]])
   Kemudian tampilkan gambar dan nilai hasil konvolusi tanpa proses clipping, dan dengan proses clipping

```python
kernel = np.array([[0, 0,-1, 1],
          [0, 2, 1, -1],
          [-1, 1, 2, 0],
          [1, -1, 0, 0]])
print(kernel)
```

menginisialisasi kernel dengan membuat kernel berukuran 4x4 dengan element-element yang telah ditentukan, kemudian menampilkan hasil dari kernel dengan fungsi print.

ZERO PADDING TANPA CLIPPING

```python
kernelTinggi, kernelLebar = kernel.shape
print('Tinggi :', kernelTinggi, 'Lebar :', kernelLebar)
```

fungsi diatas merupakan fungsi untuk mencari tinggi dan lebar pada kernel, pada library numpy terdapat fungsi shape yang berfungsi menghasilkan jumlah baris dan kolom. pada fungsi print(tinggi , lebar) berfungsi untuk menampilkan hasil tinggi dan lebar dari kernel tersebut

```python
ukuranPadding = int((kernelTinggi - 1) / 2)
```

fungsi diatas merupakan inisialisasi untuk membuat padding pada setiap sisi yang membutuhkan.

```python
lebarPadding = ((ukuranPadding, ukuranPadding+1), (ukuranPadding, ukuranPadding+1))
```

fungsi diatas digunakan untuk menentukan ukuran padding yang akan ditambahkan.

```python
paddingCipung = np.pad(cipung, lebarPadding, mode = 'constant')
```

fungsi diatas digunakan untuk menambahkan padding pada gambar cipung yang telah ditambahkan sebelumnya pada soal nomor 1

```python
zeroPadding = np.zeros((tinggi,lebar))
```

fungsi diatas merupakan fungsi yang digunakan untuk membuat sebuah array yang bernilai 0

```python
for i in range(zeroPadding.shape[0]):
    for j in range(zeroPadding.shape[1]):
        # membuat variabel fungsi untuk mengambil subset array paddingCipung dengan ukuran kernel yang sesuai
        subset = paddingCipung[i:i + kernelTinggi, j:j + kernelLebar]
        # memasukkan hasil kali subset dengan kernel pada variabel lebarSubset
        lebarSubset = subset * kernel
        # nilai rata-rata dari cipung dan kernel akan disimpan pada variabel hasilLebar
        hasilLebar = np.sum(lebarSubset)
        # fungsi untuk membulatkan nilai rata-rata hasilLebar dan disimpang pada array zeroPadding
        zeroPadding[i][j] = np.round(hasilLebar)
```

pada fungsi diatas merupakan fungsi perulangan untuk membaca tiap index pada array zeroPadding dan membuat perhitungan konvolusi zero padding.

```python
print(zeroPadding)
```

fungsi diatas merupakan fungsi untuk menampilkan hasil konvolusi dari zero padding

```python
plt.imshow(zeroPadding, cmap='gray')
```

pada fungsi diatas merupakan fungsi yang digunakan untuk menampilkan hasil dari zero padding tanpa clipping

ZERO PADDING DENGAN CLIPPING

```python
# untuk mencari nilai maximum pada pixel yang terdapat pada gambar seteleh konvolusi dilakukan
max = np.max(zeroPadding)
# untuk mencari nilai minimum pada pixel yang terdapat pada gambar seteleh konvolusi dilakukan
min = np.min(zeroPadding)
```

fungsi diatas merupakan fungsi yang digunakan untuk mencari nilai mininum dan maksimum pada gambar setelah konvolusi dilakukan

```python
clip = np.round((zeroPadding - min) / (max - min ))
```

fungsi diatas merupakan fungsi untuk melakukan clipping, round merupakan fungsi yang digunakan untuk membulatkan hasil clipping tersebut

```python
print(clip)
```

fungsi diatas merupakan fungsi untuk menampilkan hasil clipping

```python
plt.imshow(clip, cmap='gray')
```

pada fungsi diatas merupakan fungsi untuk menampilkan gambar cipung setelah dilakukan zero padding dengan clipping

TANPA PADDING DAN TANPA CLIPPING

```python
tp = np.zeros((tinggi - kernelTinggi + 1, lebar - kernelLebar + 1))
```

pada fungsi diatas merupakan fungsi untuk membuat sebuah array kosong untuk menampung hasil konvolusi tanpa padding

```python
for i in range(tp.shape[0]):
    for j in range(tp.shape[1]):
        # membuat variabel fungsi untuk mengambil subset array Cipung dengan ukuran kernel yang sesuai
        subset = cipung[i:i + kernelTinggi, j:j + kernelLebar]
        # memasukkan hasil kali subset dengan kernel pada variabel lebarSubset
        lebarSubset = subset * kernel
        # nilai rata-rata dari cipung dan kernel akan disimpan pada variabel hasilLebar
        hasilLebar = np.sum(lebarSubset)
        # fungsi untuk membulatkan nilai rata-rata yang telah ditemukan
        tp[i][j] = np.round(hasilLebar)
        
```

pada fungsi diatas merupakan fungsi perulangan untuk membaca tiap index pada array tanpa padding dan membuat perhitungan konvolusi tanpa padding

```python
print(tp)
```

fungsi diatas merupakan fungsi hasil dari konvolusi tanpa padding setelah dilakukan perhitungan

```python
plt.imshow(tp, cmap='gray')
```

pada fungsi diatas merupakan fungsi untuk menampilkan gambar cipung dengan tanpa padding dan tanpa clipping

TANPA PADDING DENGAN CLIPPING

```python
max = np.max(tp)
min = np.min(tp)
```

pada fungsi diatas merupakan fungsi untuk mencari nilai minimal dan maksimal pada gambar yang sudah dilakukan konvolusi tanpa padding

```python
clip = np.round((tp - min)/ (max - min))
```

fungsi diatas merupakan fungsi untuk melakukan clipping, round merupakan fungsi yang digunakan untuk membulatkan hasil clipping tersebut

```python
print(clip)
```

fungsi diatas merupakan fungsi untuk menampilkan hasil konvolusi clipping

```python
plt.imshow(clip, cmap='gray')
```

pada fungsi diatas merupakan fungsi untuk menampilkan gambar cipung setelah dilakukan tanpa padding dengan clipping
